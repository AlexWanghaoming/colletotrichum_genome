```python
df1 = pd.read_csv("/home/Pam/reciprocal_best_blast_hits/C413.vs.Nara.blastp.out",header=None,sep="\t")
df2 = pd.read_csv("/home/Pam/reciprocal_best_blast_hits/Nara.vs.C413.blastp.out",header=None,sep="\t")
df1.columns=["qseqid","sseqid","pident","length","mismatch","gapopen","qstart","qend","sstart","send","evalue","bitscore"]
df2.columns=["qseqid","sseqid","pident","length","mismatch","gapopen","qstart","qend","sstart","send","evalue","bitscore"]
df1.head()
```

```python
#Select rows with minimum evalue for each query seq
df3=df1.loc[df1.groupby("qseqid")["evalue"].idxmin()]
df3.head()
```

```python
#Select rows with minimum evalue for each query seq
df4=df2.loc[df2.groupby("qseqid")["evalue"].idxmin()]
df4.head()
```

```python
df3.columns=["refid1","id","pident1","length1","mismatch1","gapopen1","qstart1","qend1","sstart1","send1","evalue1","bitscore1"]
df4.columns=["id","refid2","pident2","length2","mismatch2","gapopen2","qstart2","qend2","sstart2","send2","evalue2","bitscore2"]
df3.head()
```

```python
#merge dataframes retaining only all rows all values
df5=pd.merge(df3,df4,how="outer",on="id")
df5.head()
```

```python
df6=df5[["id","refid1","refid2","evalue1","evalue2"]]
df6.head()
```

```python
#Select rows where FANref1 is FANrefid2
df7=df6[df6["refid1"]==df6["refid2"]]
df7.head()
```

```python
#list of sequences to match
query = ["CGGC5_v016292.t1"]
```
